```python
import pandas as pd
from tensorflow import keras
from tensorflow.keras import layers
from kerastuner.tuners import RandomSearch
```

```python
df=pd.read_csv("e:/my_pro_ds/deepleaning/air.txt",sep=",")
```

```python
df.head()
```

```python
#we havae to predict the pm2.5 this tells us the what is the qulity of air 
```

```python
X=df.iloc[:,:-1]#independent variables
Y=df.iloc[:,-1]# dependent variables
```

## what are hyper -parameters
#### 1)howmany number of hidden layer
#### 2. howmany no.of neurons we should have in hidden layer
#### 3.learning rate

```python
def build_model(hp):
    model=keras.Sequential()
    for i in range(hp.Int("num_layers",2,20)):
        model.add(layers.Dense(units=hp.Int("units_"+str(i),
                                           min_value=32,
                                           max_value=512,
                                           step=32),
                              activation="relu"))
    model.add(layers.Dense(1,activation="linear"))
    model.compile(
        optimizer=keras.optimizers.Adam(hp.Choice("learning_rate",[1e-2,1e-3,1e-4])),
        loss="mean_absolute_error",
        metrics=["mean_absolute_error"])
    return model
```

```python
tuner=RandomSearch(
        build_model,
        objective="val_mean_absolute_error",
        max_trials=5,
        executions_per_trial=3,
        directory="narendra33",
        project_name="ai qulity Index")
#objective means"whenever we traing the modelinside that oneparameter we get val_mean_absolute_error
```

```python
tuner.search_space_summary()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=7)
```

```python
Y_test
```

```python
tuner.search(X_train,Y_train,epochs=2,validation_data=(X_test,Y_test))
```

```python
tuner.results_summary()
```
